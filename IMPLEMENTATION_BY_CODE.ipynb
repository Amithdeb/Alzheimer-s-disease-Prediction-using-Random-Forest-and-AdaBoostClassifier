```python
data_path =  '/content/drive/MyDrive/ASSIGNMENT-3/DARWIN.csv'
df = pd.read_csv(data_path)
df
```

```python
final_data = df.drop(['ID'], axis=1)
final_data['class']=pd.factorize(final_data['class'])[0]
final_data
```

```python
X=final_data.drop(['class'], axis=1)
X
Y=final_data['class']
Y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=41)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
clf1= RandomForestClassifier(n_estimators=10)
clf2= AdaBoostClassifier(n_estimators=100, algorithm='SAMME', random_state=0)
```

```python
clf1=clf1.fit(x_train,y_train)
clf2=clf2.fit(x_train,y_train)
```

```python
clf1_pred=clf1.predict(x_test)
clf2_pred=clf2.predict(x_test)
```

```python
from sklearn import metrics
print('RF ACCURACY', metrics.accuracy_score(y_test,clf1_pred))
print('AB ACCURACY', metrics.accuracy_score(y_test,clf2_pred))
```
